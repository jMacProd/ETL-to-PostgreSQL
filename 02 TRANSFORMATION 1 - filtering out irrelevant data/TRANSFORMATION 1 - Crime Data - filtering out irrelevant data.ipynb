```python
#filter out unneccessary suburbs
# so first read in Suburb_postcard file
suburbs_postcodes_path = "../Raw Data/melbourbe_suburbs_postcode.csv"
```

```python
suburb_postcodes_DF = pd.read_csv(suburbs_postcodes_path)
suburb_postcodes_DF
```

```python
crime_suburb_merge = pd.merge(suburb_postcodes_DF, crimerateDF, on="postcode",  how ="left")
crime_suburb_merge
```

```python
#Check unique values in suburb
Checksyburbunique = crime_suburb_merge["suburb"].unique()
Checksyburbunique.sort()
len(Checksyburbunique)
```

```python
#Check unique values in state
stateeunique = crime_suburb_merge["state"].unique()
stateeunique.sort()
print(stateeunique)
len(stateeunique)
```

```python
#Drop unneccesary columns
crime_suburb = crime_suburb_merge[['Year', 'Year ending', 'Local Government Area', 'postcode',
       'Suburb/Town Name', 'Offence Division', 'Offence Subdivision',
       'Offence Subgroup', 'Incidents Recorded']]
crime_suburb
```
